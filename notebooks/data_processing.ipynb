```python
import pandas as pd

# RAW data
data = {
    "order_id": [1, 2, 3, 4],
    "customer_id": [101, 102, 101, 103],
    "product": ["Laptop", "Mouse", "Keyboard", "Monitor"],
    "amount": ["1200", "25", "75", "300"],
    "date": ["2024-01-05", "2024-01-06", "2024-01-07", "2024-01-08"]
}

df_raw = pd.DataFrame(data)

print("RAW DATA:")
display(df_raw)
```

```python
# PROCESSED layer
df_processed = df_raw.copy()

df_processed["amount"] = df_processed["amount"].astype(float)
df_processed["date"] = pd.to_datetime(df_processed["date"])

print("PROCESSED DATA:")
display(df_processed)
```

```python
# CURATED layer (aggregated for analysis)
df_curated = df_processed.groupby("customer_id")["amount"].sum().reset_index()

print("CURATED DATA:")
display(df_curated)
```
